```python
dataP2 = dataP.iloc[:, [2,10,11]].dropna(axis=0).copy() 
```

```python
dataP2 # 필요한 데이터만 추출
```

```python
dataP2["confirmed_date"] = dataP2["confirmed_date"].astype('datetime64[ns]') # 타입변환
dataP2["released_date"] = dataP2["released_date"].astype('datetime64[ns]')
```

```python
dataP2['days'] = dataP2['released_date'] - dataP2['confirmed_date'] # days 파생변수 추가
dataP2
```

```python
result = [] #문제발생 - days가 int도, object도 아니기때문에 공백을 넣어 object로 일단 전환
for x in dataP2['days']:
    result.append(x + " ")
dataP2['days'] = result
dataP2['days']
```

```python
dataP2['days'] = dataP2['days'].astype('str')
```

```python
dataP2['days'][0]
```

```python
result2 = [] # 공백을 기준으로 스플릿 처리 - 앞 숫자만 추출
for x in dataP2['days']:
    result2.append(x.split(' ')[0])
result2
```

```python
dataP2['days'] =result2
dataP2['days']
```

```python
dataP2['days'] = dataP2['days'].astype('int64') # 추출한 숫자를 int로 변환시켜줌
dataP2
```

```python
dataP2.groupby('age').count() #인원수 확인을 위해 써봄
```

```python
dataP2.groupby('age').mean() # 전반적으로 연령이 높아질수록 완치기간이 길어진다.
```
